```python
# pm2.5
pm = pd.read_csv("./data/pm25.csv", sep="\t")
pm.dropna(inplace=True)
pm["fips"] = pm["fips"].astype(int)
pm["fips"] = pm["fips"].apply("{0:0>5}".format)
pm["fips"] = pm["fips"].astype(str)


# census demographic
demographic = pd.read_csv("./data/demographic.csv")
demographic.dropna(inplace=True)
demographic["fips"] = demographic["fips"].astype(int)
demographic["fips"] = demographic["fips"].apply("{0:0>5}".format)
demographic["fips"] = demographic["fips"].astype(str)
demographic = demographic.drop(columns=["Unnamed: 0"])

# voting preferences
# voting = pd.read_csv("./data/voting.csv")
# voting = voting.rename({"FIPS": "fips"}, axis=1)
# voting["fips"] = voting["fips"].astype(str).str[:-2]
# voting["fips"] = voting["fips"].astype(str).str.pad(5, fillchar="0")
# voting["proportion"] = voting["candidatevotes"] / voting["totalvotes"]
# voting = voting.pivot(index=["state", "county", "year", "fips"], columns="party", values="proportion").add_prefix("pct_").reset_index()
# voting.drop(["pct_nan", "pct_green"], axis=1, inplace=True)
# voting.dropna(inplace=True)

# cms data
# cms = pd.read_csv("./data/cms_county.csv")
# cms["fips"] = cms["fips"].astype(str).str.pad(5, fillchar="0")
# cms = cms.drop(columns=["Unnamed: 0"])

# chronic conditions
chronic = pd.read_csv("./data/chronic.csv")
chronic.drop(["State", "County"], axis=1, inplace=True)
chronic = chronic.apply(lambda x: pd.to_numeric(x, errors="coerce")).fillna(0)
chronic["fips"] = chronic["fips"].astype(str).str.pad(5, fillchar="0")
chronic.columns = chronic.columns.str.replace(" ", "_")
chronic = chronic.drop(columns=["Unnamed:_0"])
```
